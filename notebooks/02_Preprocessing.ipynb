```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
from src.data.loader import DataLoader
from src.preprocessing.missing_value_handler import MissingValueHandler
```

```python
loader = DataLoader()

df = loader.load()

print(df.shape)
```

```python
handler = MissingValueHandler(
    strategy="median"
)

summary = handler.summary(df)

print(summary)
```

```python
clean_df = handler.fit_transform(df)

print(clean_df.shape)
```

```python
from src.preprocessing.missing_value_handler import (
    MissingValueHandler,
)

handler = MissingValueHandler(
    strategy="median"
)

summary = handler.summary(df)

print(summary)
```

```python
from src.preprocessing.duplicate_handler import DuplicateHandler

duplicate_handler = DuplicateHandler()

summary = duplicate_handler.summary(df)

print(summary)
```

```python
from src.preprocessing.outlier_handler import OutlierHandler

handler = OutlierHandler(
    strategy="ignore"
)

summary = handler.summary(df)

print(summary["outliers"]["Amount"])
```

```python
from src.preprocessing.scaler import FeatureScaler

scaler = FeatureScaler(
    strategy="standard"
)

scaled_df = scaler.fit_transform(df)

print(scaler.summary())

scaled_df.head()
```

```python
from src.preprocessing.train_test_splitter import (
    TrainTestSplitter,
)

splitter = TrainTestSplitter()

X_train, X_test, y_train, y_test = splitter.split(df)

print(splitter.summary())

print(X_train.shape)
print(X_test.shape)

print(y_train.value_counts())

print(y_test.value_counts())
```

```python
from src.preprocessing.imbalance_handler import (
    ImbalanceHandler,
)

handler = ImbalanceHandler(
    strategy="smote"
)

X_train_balanced, y_train_balanced = handler.fit_resample(
    X_train,
    y_train,
)

print(handler.summary())

print(y_train.value_counts())

print(y_train_balanced.value_counts())
```

```python
from src.preprocessing.preprocessing_pipeline import (
    PreprocessingPipeline,
)

pipeline = PreprocessingPipeline()

X_train, X_test, y_train, y_test = pipeline.run(df)

print(X_train.shape)
print(X_test.shape)

print(y_train.value_counts())

print(y_test.value_counts())
```
