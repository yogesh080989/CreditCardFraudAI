```python
from src.models.model_factory import ModelFactory
from src.models.model_trainer import ModelTrainer

# Initialize Factory and Trainer
factory = ModelFactory()
trainer = ModelTrainer()

# Models to train
models = [
    "logistic_regression",
    "decision_tree",
    "random_forest",
    "xgboost",
]

trained_models = {}
training_summary = []

for model_name in models:

    print("=" * 70)
    print(f"Training: {model_name}")

    # Create model
    model = factory.get_model(model_name)

    # Train model
    trained_model = trainer.train(
        model,
        X_train,
        y_train,
    )

    # Store trained model
    trained_models[model_name] = trained_model

    # Store metadata
    metadata = trainer.summary().copy()
    training_summary.append(metadata)

    print(metadata)
    print()
```

```python
import pandas as pd

training_summary_df = pd.DataFrame(training_summary)

training_summary_df
```

```python
from src.evaluation.model_evaluator import ModelEvaluator

evaluator = ModelEvaluator()

evaluation_results = []

for name, model in trained_models.items():

    report = evaluator.evaluate(
        model,
        X_test,
        y_test,
    )

    evaluation_results.append(report)

    print("=" * 70)
    print(name)
    print("=" * 70)

    print(f"Accuracy : {report['accuracy']:.6f}")
    print(f"Precision: {report['precision']:.6f}")
    print(f"Recall   : {report['recall']:.6f}")
    print(f"F1 Score : {report['f1_score']:.6f}")
    print(f"ROC AUC  : {report['roc_auc']:.6f}")

    print()
```

```python
import pandas as pd

comparison_df = pd.DataFrame([
    {
        "Model": r["model_name"],
        "Accuracy": r["accuracy"],
        "Precision": r["precision"],
        "Recall": r["recall"],
        "F1 Score": r["f1_score"],
        "ROC AUC": r["roc_auc"],
    }
    for r in evaluation_results
])

comparison_df.sort_values(
    by="F1 Score",
    ascending=False,
)
```
