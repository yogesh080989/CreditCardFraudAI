This notebook demonstrates the Explainable AI (XAI) layer developed for the Credit Card Fraud Detection system.

```python
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import sys

PROJECT_ROOT = Path("../").resolve()

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print(PROJECT_ROOT)
```

## Import Required Modules

Import all enterprise components developed during previous sprints.

```python
import joblib
import pandas as pd

from src.explainability.shap_explainer import SHAPExplainer
from src.ai.explanation_engine import ExplanationEngine
```

## Load Trained Model

Load the production-ready Random Forest model generated during Sprint 3.

```python
MODEL_PATH = "../models/RandomForestClassifier.pkl"

model = joblib.load(MODEL_PATH)

print(type(model))
```

## Load Test Dataset

Load the processed test dataset that will be used for generating AI explanations.

```python
X_test = pd.read_csv("../data/processed/X_test.csv")

print(X_test.shape)

X_test.head()
```

```python
## Initialize SHAP Explainer

explainer = SHAPExplainer(
    model=model,
    feature_names=list(X_test.columns),
    output_dir="../artifacts/shap"
)
```

```python
## Compute SHAP Values

explainer.compute_shap_values(X_test)
```

## Generate Explainability Report

Generate an explainability report for a single transaction.

This report contains:

- Prediction
- Fraud Probability
- Risk Level
- Top Contributing Features

```python
report = explainer.generate_explainability_report(
    row_index=0
)

report
```

## Initialize AI Explanation Engine

The AI Explanation Engine transforms the SHAP explainability report into business-friendly explanations suitable for fraud analysts, executives, and auditors.

```python
engine = ExplanationEngine()
```

## Generate AI Explanation

Convert the SHAP report into a structured AI explanation.

```python
ai_result = engine.generate(
    explainability_report=report,
    model_name=explainer.model_name
)
```

## Executive Summary

Display a concise summary for senior management.

```python
print(ai_result.executive_summary.headline)
print()

print(ai_result.executive_summary.one_line_summary)
print()

print(ai_result.executive_summary.recommendation)
```

## Business Summary

Business-friendly explanation describing why the transaction was classified as fraud.

```python
print(ai_result.business_summary.summary)
```

## Technical Summary

Technical explanation generated using SHAP feature attribution.

```python
print(ai_result.technical_summary.explanation)
```

## Confidence Statement

Model confidence for the generated prediction.

```python
print(ai_result.confidence_statement)
```

## Feature Contributions

Display the most influential features contributing to the fraud prediction.

```python
ai_result.top_features_dataframe()
```

## Recommendation

Recommended operational action based on the predicted fraud risk.

```python
print(ai_result.recommendation)
```

## Metadata

Display metadata generated by the AI Explanation Engine.

```python
ai_result.metadata
```
